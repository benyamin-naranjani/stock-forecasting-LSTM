```python
from nbformat import v4 as nbf

nb = nbf.new_notebook()

text = """
# Evaluation Metrics for Portfolio Performance

## 1. Total Return
**Definition:**  
The percentage change in your portfolio value from start to finish.

$$
\\text{Total Return} = \\frac{V_{\\text{end}} - V_{\\text{start}}}{V_{\\text{start}}}
$$

---

## 2. Sharpe Ratio
**Definition:**  
Risk-adjusted return. Measures how much excess return you generate per unit of volatility.

$$
\\text{Sharpe} = \\frac{R_p - R_f}{\\sigma_p}
$$

Where:  
- \( R_p \) = average portfolio return  
- \( R_f \) = risk-free rate (can be set to 0 for simplicity)  
- \( \\sigma_p \) = portfolio volatility  

---

## 3. Max Drawdown
**Definition:**  
The worst peak-to-trough decline in portfolio value.

$$
\\text{Max Drawdown} = \\max_t \\left( \\frac{\\text{Peak}(t) - \\text{Val}(t)}{\\text{Peak}(t)} \\right)
$$

---

## 4. Volatility
**Definition:**  
Standard deviation of returns, annualized.

$$
\\sigma_{\\text{annual}} = \\sigma_{\\text{period}} \\times \\sqrt{\\text{periods per year}}
$$

---

## 5. Worst Losing Streak
**Definition:**  
The longest consecutive sequence of losing periods (e.g., losing months).

---

## 6. Turnover
**Definition:**  
How much of the portfolio changes during each rebalance.

$$
\\text{Turnover} = \\frac{\\sum |w_i^{\\text{new}} - w_i^{\\text{old}}|}{2}
$$
"""

nb['cells'] = [nbf.new_markdown_cell(text)]

path = "portfolio_metrics_definitions.ipynb"
with open(path, "w", encoding="utf-8") as f:
    f.write(nbf.writes(nb))

path
```
